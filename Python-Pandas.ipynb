```python
# Applying A Function To A Row
# Find minor and adult passengers 
def is_minor(row):
    if row["Age"] < 18:
        return "Minor"
    elif row["Age"] >= 18:
        return "Adult"
    else:
        return "Unknown"

age_labels = titanic_survival.apply(is_minor, axis=1)
age_labels
```

```python
# Calculating Survival Percentage By Age Group
def is_minor(row):
    if row["Age"] < 18:
        return "Minor"
    elif row["Age"] >= 18:
        return "Adult"
    else:
        return "Unknown"

age_labels = titanic_survival.apply(is_minor, axis=1)
titanic_survival['age_labels'] = age_labels
age_group_survival = titanic_survival.pivot_table(values='Survived', index=['age_labels'])
age_group_survival
age_group_survival
```

```python
#GroupBy
import pandas as pd
bikes_data = pd.read_csv("bikes.csv", sep=';', encoding='latin1', parse_dates=['Date'], dayfirst=True, index_col='Date')
bikes_data
```

```python
berri_bikes = bikes_data[['Berri 1']].copy()
berri_bikes[:5]
```

```python
# Weekdays 0-monday, 1 Tuesday .....
berri_bikes.index.weekday
```

```python
# Add weekdays column
berri_bikes.loc[:,'weekday'] = berri_bikes.index.weekday
berri_bikes[:5]
```

```python
# No of cyclist on each day during the year
weekday_counts = berri_bikes.groupby('weekday').aggregate(sum)
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts
```

```python
# Find the mean() of cyclists by weekday
weekday_mean = berri_bikes.groupby('weekday').mean()
weekday_mean.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_mean
```

```python
# Apply all the aggregate methods on cyclist data
weekday_min = berri_bikes.groupby('weekday')
weekday_min.describe()
```

```python
# Concatenating maultiple data frames
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
pd.concat([df1,df2,df3])
```

```python
# Merging data frames
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
pd.merge(left,right,how='inner',on='key')
```

```python
# Merging on multiple keys
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                         'C': ['C0', 'C1', 'C2', 'C3'],
                         'D': ['D0', 'D1', 'D2', 'D3']})
pd.merge(left, right, on=['key1', 'key2'])
```

```python
# Outer Merging on multiple keys
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
# Right merge
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
# Left merge
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
# Join
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
left.join(right)
```

```python
left.join(right, how='outer')
```
