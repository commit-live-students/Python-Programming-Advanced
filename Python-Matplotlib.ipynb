```python
# Generate a line chart that visualizes the unemployment rates from 1948
import pandas as pd
unemployement_rate = pd.read_csv('unemployement_rate.csv')
# Convert the DATE column to datetime datatype
unemployement_rate['DATE'] = pd.to_datetime(unemployement_rate['DATE'])
data_1948 = unemployement_rate.head(12)
plt.plot(data_1948['DATE'], data_1948['VALUE'], 'r') # 'r' is the color red
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.xticks(rotation=90) # Rotate xticks 90 degree
plt.title('Unemployment Trends, 1948')
plt.show()
```

```python
# Fig and axes object
import numpy as np
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.3]) # left, bottom, width, height (range 0 to 1)

x = np.array([0, 1, 2, 3, 4, 5])
y = x ** 2
# Plot on that set of axes
axes.plot(x, y, 'b') # b - blue
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
plt.show()
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
plt.show()
```

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
plt.show()
```

```python
# Exercise : Create 2 line subplots in a 2 row by 1 column layout
import pandas as pd
unemployement_rate = pd.read_csv('unemployement_rate.csv')
# Convert the DATE column to datetime datatype
unemployement_rate['DATE'] = pd.to_datetime(unemployement_rate['DATE'])

# 2 row by 1 column layout
fig, axes = plt.subplots(nrows=2, ncols=1)
# In the top subplot, plot the data from 1948.
axes[0].plot(unemployement_rate[0:12]['DATE'], unemployement_rate[0:12]['VALUE'], 'r')
# In the bottom subplot, plot the data from 1949.
axes[1].plot(unemployement_rate[12:24]['DATE'], unemployement_rate[12:24]['VALUE'], 'r')
plt.show()
```

```python
# Exercise : Create 5 line subplots in a 5 row by 1 column layout
import pandas as pd
unemployement_rate = pd.read_csv('unemployement_rate.csv')
# Convert the DATE column to datetime datatype
unemployement_rate['DATE'] = pd.to_datetime(unemployement_rate['DATE'])

# 2 row by 1 column layout
fig, axes = plt.subplots(6, 1, figsize=(12,12))

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    axes[i].set_title(str(1948+i))
    axes[i].plot(unemployement_rate[start_index:end_index]['DATE'], unemployement_rate[start_index:end_index]['VALUE'], 'r')
    
plt.show()
```

```python
# Exercise : Generate 2 line charts in the base subplot using the MONTH column for the x-axis 
import pandas as pd
unemployement_rate = pd.read_csv('unemployement_rate.csv')
# Convert the DATE column to datetime datatype
unemployement_rate['DATE'] = pd.to_datetime(unemployement_rate['DATE'])
unemployement_rate['MONTH'] = unemployement_rate['DATE'].dt.month
fig = plt.figure(figsize=(6,3))

plt.plot(unemployement_rate[0:12]['MONTH'], unemployement_rate[0:12]['VALUE'], c='red')
plt.plot(unemployement_rate[12:24]['MONTH'], unemployement_rate[12:24]['VALUE'], c='blue')

plt.show()
```

```python
# Exercise : Generate 2 line charts in the base subplot using the MONTH column for the x-axis, add Legend 1948, 1949
import pandas as pd
unemployement_rate = pd.read_csv('unemployement_rate.csv')
# Convert the DATE column to datetime datatype
unemployement_rate['DATE'] = pd.to_datetime(unemployement_rate['DATE'])
unemployement_rate['MONTH'] = unemployement_rate['DATE'].dt.month
fig = plt.figure(figsize=(6,3))

plt.plot(unemployement_rate[0:12]['MONTH'], unemployement_rate[0:12]['VALUE'], c='red', label="1948")
plt.plot(unemployement_rate[12:24]['MONTH'], unemployement_rate[12:24]['VALUE'], c='blue', label="1949")
plt.legend()
plt.show()
```

```python
# Plot range
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
plt.show()
```

```python
# Exercise: Generate a bar plot for Fandango review

import pandas as pd
import numpy as np
reviews = pd.read_csv('fandango_scores.csv')
cols = ['FILM', 'RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
norm_reviews = reviews[cols]
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']

bar_heights = norm_reviews.mean(axis=0)
bar_positions = np.arange(5) + 0.75
tick_positions = range(1,6)
fig, ax = plt.subplots()
ax.bar(bar_positions, bar_heights, 0.5)
ax.set_xticks(tick_positions)
ax.set_xticklabels(num_cols, rotation=90)

ax.set_xlabel('Rating Source');
ax.set_ylabel('Average Rating');
ax.set_title('Average User Rating For Avengers: Age of Ultron (2015)')
plt.show()
```
